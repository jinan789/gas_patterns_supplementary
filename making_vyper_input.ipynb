```python
from making_table_helper import *
import random
import os
import json
import re
from collections import defaultdict

# File path to the text file containing hashes
file_folder = './etherscan_cache/vyper_recent_contracts/'
all_vyper_hashes = []

for single_file in os.listdir(file_folder):
    if single_file.startswith('.'):
        continue
    file_path = os.path.join(file_folder, single_file)
    
    # Dictionary to store hash occurrences
    hash_occurrences = defaultdict(int)

    # Regular expression pattern to match hexadecimal hashes
    # This pattern matches strings that start with "0x" or "Q_0x"
    # followed by one or more hexadecimal characters
    hash_pattern = re.compile(r'0x[0-9a-fA-F]+')

    # Read the file and count hash occurrences
    with open(file_path, 'r') as file:
        for line in file:
            # Find all matches of the hash pattern in the line
            matches = hash_pattern.findall(line)
            # Update the occurrences of each hash in the dictionary
            for match in matches:
                # Concatenate the match parts to form the full hash
                full_hash = ''.join(match)
                hash_occurrences[full_hash.lower()] += 1

    # Convert defaultdict to a regular dict for the final result
    hash_occurrences = dict(hash_occurrences)

    hash_occurrences = {k:v for k,v in hash_occurrences.items() if v == 5}

    print(len(hash_occurrences))
    all_vyper_hashes.extend(list(hash_occurrences.keys()))
#     # Print the dictionary mapping hash to the number of occurrences
#     for hash_code, count in hash_occurrences.items():
#         print(f"Hash: {hash_code}, Occurrences: {count}")
```

```python
def write_to_file(contents, path):
    with open(path, 'w') as f:
        f.write(contents)
    
def read_from_file(path):
    with open(path) as f:
        return ''.join(f.readlines())
```

```python
path_to_codes_map = dict()
for p in contract_names_for_filtering:
    print(p)
    path_to_codes_map[p] = json.loads(read_from_file(p))['result'][0]['SourceCode']
```

```python
function_counts = {}
for k in cur_filtered_list_to_title:
    cur_functions = []
    for line in cur_filtered_list_to_title[k].split('\n'):
        if line.startswith('def '):
            func_part = line.split('(')[0]
            assert func_part.startswith('def ')
            func_part = func_part[4:]
            cur_functions.append(func_part)
            
    for func_part in cur_functions:
        if func_part not in function_counts:
            function_counts[func_part] = 0
        function_counts[func_part] += 1
        
# print(cur_filtered_list['./etherscan_cache/getsourcecode/0xe2554caa829acace7042a3808275dfdb97a6187c.txt'])

```

```python
salient_functions = []
for k in function_counts:
    if function_counts[k] == 1:
        salient_functions.append(k)
```

```python
len(salient_functions)
```

```python
# need to take the top 10 percent of the functions, by length.

function_to_file_name = dict()
function_to_contents = dict()

for k in cur_filtered_list_to_title:    
    is_in_func = False
    cur_accumulated_lines = []
    cur_func_part = None
    
    for line in cur_filtered_list_to_title[k].split('\n'):
        if is_in_func and not line.startswith('    '):
            if cur_func_part in salient_functions:
                assert cur_func_part not in function_to_contents
                function_to_contents[cur_func_part] = '\n'.join(cur_accumulated_lines)
                function_to_file_name[cur_func_part] = k
            is_in_func = False
            
            
        if line.startswith('def '):
            # clear previous function
            if is_in_func:
                if cur_func_part in salient_functions:
                    assert cur_func_part not in function_to_contents
                    function_to_contents[cur_func_part] = '\n'.join(cur_accumulated_lines)
                    function_to_file_name[cur_func_part] = k
                
                
            func_part = line.split('(')[0]
            assert func_part.startswith('def ')
            func_part = func_part[4:]
            
            cur_accumulated_lines = []
            cur_func_part = func_part
            is_in_func = True
            
        cur_accumulated_lines.append(line)
        
    if is_in_func:
        if cur_func_part in salient_functions:
            assert cur_func_part not in function_to_contents
            function_to_contents[cur_func_part] = '\n'.join(cur_accumulated_lines)
            function_to_file_name[cur_func_part] = k
```

```python
# NEWEST VERSION

m = 5500
prefix = read_from_file('gas_new_pattern_prompts_vyper/' + 'new_pattern_discovery' + '.txt')
postfix = read_from_file('gas_new_pattern_prompts_vyper/' +  'new_pattern_discovery_postfix.txt')


cur_filtered_list_to_title_salient_final_version = dict()

for k in cur_filtered_list_to_title_salient:
    cur_func_name_to_function_d = dict()
    
    
    is_in_func = False
    cur_accumulated_lines = []
    cur_func_part = None
    
    for line in [i for i in cur_filtered_list_to_title[k].split('\n') if i.strip() != '']:
        if line.startswith('def '):
            # clear previous function
            if is_in_func:
                assert cur_func_part not in cur_func_name_to_function_d
                cur_func_name_to_function_d[cur_func_part] = '\n'.join(cur_accumulated_lines)
                
                
            func_part = line.split('(')[0]
            assert func_part.startswith('def ')
            func_part = func_part[4:]
            
            cur_accumulated_lines = []
            cur_func_part = func_part
            is_in_func = True
            
        cur_accumulated_lines.append(line)
        
    cur_func_name_to_function_pairs = sorted(cur_func_name_to_function_d.items(), key = lambda x: -len(x[1]))
    
    cur_content = ''
    for pair in cur_func_name_to_function_pairs:
        cur_content += pair[1]
        if len(cur_content) >= m:
            print(1)
            break
            
            
    cur_content = prefix + cur_content + postfix
    hashcode = k.split('/')[-1]
    cur_filtered_list_to_title_salient_final_version[hashcode] = cur_content
    
            
            
            
            
            
        
```
