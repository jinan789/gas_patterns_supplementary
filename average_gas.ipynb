```python
D = []
M = []
for k in pattern_to_gas_cost:
    cur_d = pattern_to_gas_cost[k]
    d_before = int(cur_d['deployment_execution_before'])
    d_after = int(cur_d['deployment_execution_after'])
    m_before = int(cur_d['message_call_execution_before'])
    m_after = int(cur_d['message_call_execution_after'])
    D.append((d_before - d_after) / d_before * 100)
    M.append((m_before - m_after) / m_before * 100)
    print((d_before - d_after) / d_before * 100, (m_before - m_after) / m_before * 100)
#     print()
```
